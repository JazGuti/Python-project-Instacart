## 5. Determining a Suitable way to combine

The most suitable way to combine both dataframes is with a merge with column 'product_id'. 

```python
# Creating Dataframe with merged df_ords_prods_combined and df_prods
df_ords_prods_merge = df_ords_prods_combined.merge(df_prods, on = ['product_id'])
```

```python
# Checking outcome
df_ords_prods_merge.head()
```

```python
# Dropping unnamed columns
df_ords_prods_merge = df_ords_prods_merge.drop(columns = ['Unnamed: 0_x', 'Unnamed: 0.1', 'Unnamed: 0_y'])
```
